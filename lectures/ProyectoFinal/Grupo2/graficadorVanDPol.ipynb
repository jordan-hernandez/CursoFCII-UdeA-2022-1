Codigo para graficar los datos obtenidos de la ejecucion del programa en c++

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Variamos $\mu$ de 0 a 10 con las siguientes condiciones

    // void disparo(a, b, alpha, beta, N, TOL, M, mu)
    
    ejemplo.disparo(1, 2, 0, 1, 50, 10e-2, 10000, 0);

```python
data = pd.read_csv("graficaVanderPol0.000000.txt",names=["x","y","v"])
data2 = pd.read_csv("graficaVanderPol0.200000.txt",names=["x","y","v"])
data3 = pd.read_csv("graficaVanderPol0.500000.txt",names=["x","y","v"])
data4 = pd.read_csv("graficaVanderPol1.000000.txt",names=["x","y","v"])
data5 = pd.read_csv("graficaVanderPol2.000000.txt",names=["x","y","v"])
data6 = pd.read_csv("graficaVanderPol4.000000.txt",names=["x","y","v"])
data7 = pd.read_csv("graficaVanderPol6.000000.txt",names=["x","y","v"])
data8 = pd.read_csv("graficaVanderPol8.000000.txt",names=["x","y","v"])
data9 = pd.read_csv("graficaVanderPol10.000000.txt",names=["x","y","v"])

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(10)
plt.grid()

plt.plot(data["x"],data["y"],label="$\mu=0$")
plt.plot(data2["x"],data2["y"],label="$\mu=0.2$")
plt.plot(data3["x"],data3["y"],label="$\mu=0.5$")
plt.plot(data4["x"],data4["y"],label="$\mu=1$")
plt.plot(data5["x"],data5["y"],label="$\mu=2$")
plt.plot(data6["x"],data6["y"],label="$\mu=4$")
plt.plot(data7["x"],data7["y"],label="$\mu=6$")
plt.plot(data8["x"],data8["y"],label="$\mu=8$")
plt.plot(data9["x"],data9["y"],label="$\mu=10$")

plt.legend()
plt.xlabel("time",fontsize=15)
plt.ylabel("position",fontsize=15)

plt.rc("xtick",labelsize=12)
plt.rc("ytick",labelsize=12)
plt.show()
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(10)
plt.grid()

plt.plot(data["v"],data["y"],label="$\mu=0$")
plt.plot(data2["v"],data2["y"],label="$\mu=0.2$")
plt.plot(data3["v"],data3["y"],label="$\mu=0.5$")
plt.plot(data4["v"],data4["y"],label="$\mu=1$")
plt.plot(data5["v"],data5["y"],label="$\mu=2$")
plt.plot(data6["v"],data6["y"],label="$\mu=4$")
plt.plot(data7["v"],data7["y"],label="$\mu=6$")
plt.plot(data8["v"],data8["y"],label="$\mu=8$")
plt.plot(data9["v"],data9["y"],label="$\mu=10$")

plt.legend()
plt.xlabel("velocity",fontsize=15)
plt.ylabel("position",fontsize=15)

plt.rc("xtick",labelsize=12)
plt.rc("ytick",labelsize=12)
plt.show()
```
